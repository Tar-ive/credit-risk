```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(15, 10))

plt.subplot(2, 3, 1)
sns.histplot(X['X1'], bins=30, kde=True)
plt.title('Distribution of X1 (Amount of Credit Given)')
plt.xlabel('Amount of Credit')
plt.ylabel('Count')

plt.subplot(2, 3, 2)
sns.histplot(X['X2'], bins=10, kde=False)
plt.title('Distribution of X2 (Gender)')
plt.xlabel('Gender (1 = male, 2 = female)')
plt.ylabel('Count')

plt.subplot(2, 3, 3)
sns.histplot(X['X3'], bins=10, kde=False)
plt.title('Distribution of X3 (Education)')
plt.xlabel('Education Level (1 = graduate school; 2 = university; 3 = high school; 4 = others)')
plt.ylabel('Count')

plt.subplot(2, 3, 4)
sns.histplot(X['X4'], bins=10, kde=False)
plt.title('Distribution of X4 (Marital Status)')
plt.xlabel('Marital Status (1 = married, 2 = single, 3 = others)')
plt.ylabel('Count')

plt.subplot(2, 3, 5)
sns.histplot(X['X5'], bins=20, kde=True)
plt.title('Distribution of X5 (Age)')
plt.xlabel('Age')
plt.ylabel('Count')

plt.tight_layout()
plt.show()
```

```python
age_bins = [20, 30, 40, 50, 60, 100]
age_labels = ['21-30', '31-40', '41-50', '51-60', '60+']

X['AgeGroup'] = pd.cut(X['X5'], bins=age_bins, labels=age_labels)
average_x1_by_age = X.groupby('AgeGroup')['X1'].mean().reset_index()

plt.figure(figsize=(10, 6))
sns.barplot(x='AgeGroup', y='X1', data=average_x1_by_age, order=age_labels)
plt.title('Average X1 by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Average Amount of Credit (X1)')
plt.show()
```
